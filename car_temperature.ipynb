## 🔹 Project Overview
This project is a **car temperature monitoring system.** It simulates reading the temperature inside a car and sends an alert when the temperature becomes too high. The system improves safety by notifying the user in real time.



# 🎯 Objective

The main goal of this project is to:

Monitor the car’s temperature continuously
Detect when the temperature exceeds a safe limit
Send an instant notification to the user using an external service (Pushover)

This helps prevent dangerous situations caused by high temperatures.

# ⚙️ How It Works

The system generates a temperature value (simulated using random numbers)
It checks if the temperature is higher than a specific threshold (e.g., 100°C)
If the temperature is high, it sends a notification to the user via Pushover API
The process repeats continuously at fixed time intervals

```python
import requests
import random
import time

# بيانات Pushover
USER_KEY = "ugdef6zccci2y9s1qn31z8pmhhajzc"
API_TOKEN = "ajzqd2dgnmm2caqn7yn8qzzrurrqe9"

# إرسال تنبيه
def send_alert(message):
    url = "https://api.pushover.net/1/messages.json"
    data = {
        "token": API_TOKEN,
        "user": USER_KEY,
        "message": message
    }
    requests.post(url, data=data)

# قراءة الحرارة
def read_temperature():
    return random.randint(70, 120)

# system
def car_agent():
    temp = read_temperature()
    print(f"Current Temperature: {temp}")

    if temp > 100:
        print("تم إرسال تنبيه")
        send_alert(f" حرارة السيارة مرتفعه  {temp}°C")
    else:
        print(" طبيعي")

# تشغيل
for i in range(5):
    car_agent()
    time.sleep(5)
```
